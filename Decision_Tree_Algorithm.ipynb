### **Prediction using Decision Tree Algorithm**

● Creating the Decision Tree classifier and visualizing it graphically.

● The purpose is if we feed any new data to this classifier, it would be able to predict the right class accordingly.

```python
# importing the libraries
import numpy as np, pandas as pd, seaborn as sns, matplotlib.pyplot as plt
import warnings
warnings.filterwarnings('ignore')
from sklearn.metrics import accuracy_score, confusion_matrix
from sklearn.model_selection import train_test_split
from sklearn import tree
```

```python
df_iris = pd.read_csv('https://raw.githubusercontent.com/AumJain2305/DS-303/main/Iris.csv')
df_iris.head()
```

```python
df_iris.shape
```

```python
X= df_iris.drop(["Id","Species"],axis=1)
y= df_iris['Species']
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.25)
```

```python
from sklearn.tree import DecisionTreeClassifier
dtc = DecisionTreeClassifier()
```

```python
new_model_1 = dtc.fit(X_train,y_train)
```

```python
Y_pred = dtc.predict(X_test)
```

```python
df2 = pd.DataFrame({"Actual":y_test,"Predicted":Y_pred})
df2.head(100)
```

```python
accuracy_score(y_test,Y_pred)
```
